#Atividade I

```python
import requests
```

```python
def fetch_dados_paises():
  resposta = requests.get(
      'https://restcountries.com/v3.1/all?fields=name,population,region'
  )

  dados = resposta.json()

  dados_tratados = []

  for pais in dados:
    dados_tratados.append({
        'nome': pais.get('name').get('common'),
        'populacao': pais.get('population'),
        'regiao': pais.get('region')
    })

  return dados_tratados
```

```python
print(fetch_dados_paises())
```

# Atividade II

```python
def fetch_pais_por_nome(nome):
  resposta = requests.get(
      f'https://restcountries.com/v3.1/name/{nome}'
  )
  return resposta.json()[0]
```

```python
print(fetch_pais_por_nome('Senegal'))
```

```python
def fetch_codigo_iso_moeda_por_pais(nome):
  dados = fetch_pais_por_nome(nome)
  return list(dados.get('currencies').keys())[0]
```

```python
fetch_codigo_iso_moeda_por_pais('Monaco')
```

```python
def fetch_idioma_por_pais(nome):
  dados = fetch_pais_por_nome(nome)
  return dados.get('languages')
```

```python
fetch_idioma_por_pais('Australia')
```

```python
def fecth_pais_africano_idependente():
  resposta = requests.get(
      'https://restcountries.com/v3.1/independent?status=true'
  )

  dados = resposta.json()

  return [pais for pais in dados if pais.get('region') == 'Africa']
```

```python
print(fecth_pais_africano_idependente())
```
